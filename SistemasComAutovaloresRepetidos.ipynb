### Os exercícios são extraídos do livro do Boyce $\&$ DiPrima $10^{th}$ edition

## Exercício 1 da página 436 do livro do Boyce, $10^{th}$ edition

### Descubra a solução geral para o seguinte sistema de equações diferenciais ordinárias de primeira ordem:

$\dot{\mathbf{x}} = \begin{pmatrix}
3 & -4 \\
1 & -1
\end{pmatrix} \mathbf{x}$.

A solução geral é dada por:

$\mathbf{x}(t) = c_1 \mathbf{v}_1 e^{\lambda_1 t} + c_2 \mathbf{v}_2 e^{\lambda_2 t}$, onde $\lambda_1 e \lambda_2$ são os autovalores do sistema e $\mathbf{v}_1$ e $\mathbf{v}_2$ são os autovetores. Vamos provar que essa é uma solução do sistema.

Se $\mathbf{x} = \mathbf{v}e^{\lambda t}$ é uma solução de $\dot{\mathbf{x}} = \mathbf{Ax}$, 
$\dot{\mathbf{x}} = \lambda \mathbf{v}e^{\lambda t} =  \mathbf{Ax} = \mathbf{Av}e^{\lambda t} 
\longrightarrow \lambda \mathbf{v}e^{\lambda t} = \mathbf{Av}e^{\lambda t}$. Como $e^{\lambda t}$ jamais será $0.0$, podemos dividir ambos os lados da equação por $e^{\lambda t}$ e obteremos $\mathbf{Av} = \lambda \mathbf{v}$.

Calculando autovalores e autovetores:

$\mathbf{Ax} = \lambda\mathbf{x} \longrightarrow |(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = 0$

$|(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = \begin{vmatrix}
3 - \lambda & -4 \\
1 & -1 - \lambda
\end{vmatrix}$

A equação característica é:

$\lambda^2 - 2\lambda + 1 = 0 \longrightarrow \lambda = 1$. Só existe um autovalor.

Vamos calcular o autovetor associado a este autovalor.

$\begin{pmatrix}
3 & -4 \\
1 & -1
\end{pmatrix} 
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = 1
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} \longrightarrow \mathbf{v} =
\begin{pmatrix}
2 \\
1
\end{pmatrix}
$
A primeira parte da solução geral é:

$\mathbf{x}_1(t) = c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{t}$

Vamos calcular agora a segunda parte da equação geral, supondo que seja da seguinte forma:

$\mathbf{x}(t) = \mathbf{\xi} te^{\lambda t} + \mathbf{\eta} e^{\lambda t}$. Como $\lambda = 1$, temos
$\mathbf{x}(t) = \mathbf{\xi} te^{t} + \mathbf{\eta} e^{t}$.

$\dot{\mathbf{x}} = \mathbf{Ax}$

$\mathbf{\xi} e^{t} + \mathbf{\xi} te^{t} + \mathbf{\eta} e^{t} = \mathbf{A \xi}t e^{t} + \mathbf{A \eta}e^{t}$

$ \mathbf{A \xi}t e^{t} = \mathbf{\xi} te^{t} \longrightarrow \mathbf{A \xi} = \mathbf{\xi}$. O vetor $\mathbf{\xi}$ é o autovetor de $\mathbf{A}$, já calculado anteriormente. 

$\mathbf{\xi} = \begin{pmatrix}
2  \\
1 
\end{pmatrix}$.

$\mathbf{\xi} e^{t} + \mathbf{\eta} e^{t} = \mathbf{A \eta}e^{t} \longrightarrow \big(\mathbf{A - I} \big)\mathbf{\eta} = \mathbf{\xi}$

$\big(\mathbf{A - I} \big)\mathbf{\eta} = \mathbf{\xi} \longrightarrow$
$\begin{pmatrix}
3 - 1& -4 \\
1 & -1 - 1
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix} \longrightarrow$
$\begin{pmatrix}
2& -4 \\
1 & -2
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix}$

Resolvendo a equação, temos que $\eta_1 - 2 \eta_2 = 1 \longrightarrow \eta_1 = 2 \eta_2 + 1$.

Se considerarmos $\eta_2$ como uma variável livre, com valor $k$, teremos:

$\mathbf{\eta} = \begin{pmatrix}
2k + 1 \\
k
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix}k + \begin{pmatrix}
1 \\
0
\end{pmatrix}$.

Dos cálculos anteriores, temos:

$\mathbf{x}_2(t) = c_2 \Big( \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{t} + \begin{pmatrix}
2 \\
1
\end{pmatrix} k e^{t} + \begin{pmatrix}
1 \\
0
\end{pmatrix}e^{t} \Big)$

$\mathbf{x}(t) = \mathbf{x}_1(t) + \mathbf{x}_2(t) =  c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{t} + c_2 \Big(  \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{t} + \begin{pmatrix}
2 \\
1
\end{pmatrix} k e^{t} + \begin{pmatrix}
1 \\
0
\end{pmatrix}e^{t} \Big) = $

$\mathbf{x}(t) = c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{t} + c_2 \Big(  \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{t} +  \begin{pmatrix}
1 \\
0
\end{pmatrix}e^{t} \Big) $

## Exercício 2 da página 436 do livro do Boyce, $10^{th}$ edition

### Descubra a solução geral para o seguinte sistema de equações diferenciais ordinárias de primeira ordem:

$\dot{\mathbf{x}} = \begin{pmatrix}
4 & -2 \\
8 & -4
\end{pmatrix} \mathbf{x}$.

A solução geral é dada por:

$\mathbf{x}(t) = c_1 \mathbf{v}_1 e^{\lambda_1 t} + c_2 \mathbf{v}_2 e^{\lambda_2 t}$, onde $\lambda_1 e \lambda_2$ são os autovalores do sistema e $\mathbf{v}_1$ e $\mathbf{v}_2$ são os autovetores. Vamos provar que essa é uma solução do sistema.

Se $\mathbf{x} = \mathbf{v}e^{\lambda t}$ é uma solução de $\dot{\mathbf{x}} = \mathbf{Ax}$, 
$\dot{\mathbf{x}} = \lambda \mathbf{v}e^{\lambda t} =  \mathbf{Ax} = \mathbf{Av}e^{\lambda t} 
\longrightarrow \lambda \mathbf{v}e^{\lambda t} = \mathbf{Av}e^{\lambda t}$. Como $e^{\lambda t}$ jamais será $0.0$, podemos dividir ambos os lados da equação por $e^{\lambda t}$ e obteremos $\mathbf{Av} = \lambda \mathbf{v}$.

Calculando autovalores e autovetores:

$\mathbf{Ax} = \lambda\mathbf{x} \longrightarrow |(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = 0$

$|(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = \begin{vmatrix}
4 - \lambda & -2 \\
8 & -4 - \lambda
\end{vmatrix}$

A equação característica é:

$\lambda^2 = 0 \longrightarrow \lambda = 0$. Só existe um autovalor.

Vamos calcular o autovetor associado a este autovalor.

$\begin{pmatrix}
4 & -2 \\
8 & -4
\end{pmatrix} 
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = 0
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} \longrightarrow \mathbf{v} =
\begin{pmatrix}
1 \\
2
\end{pmatrix}
$
A primeira parte da solução geral é:

$\mathbf{x}_1(t) = c_1 \begin{pmatrix}
1  \\
2 
\end{pmatrix}e^{0t} = c_1\begin{pmatrix}
1  \\
2 
\end{pmatrix} $

Vamos calcular agora a segunda parte da equação geral, supondo que seja da seguinte forma:

$\mathbf{x}(t) = \mathbf{\xi} te^{\lambda t} + \mathbf{\eta} e^{\lambda t}$. Como $\lambda = 0$, temos
$\mathbf{x}(t) = \mathbf{\xi} t + \mathbf{\eta}$.

$\dot{\mathbf{x}} = \mathbf{Ax}$

$\mathbf{\xi} = \mathbf{A \xi}t + \mathbf{A \eta}$

$ \mathbf{A \xi}t = 0 \mathbf{\xi}$. O vetor $\mathbf{\xi}$ é o autovetor de $\mathbf{A}$, já calculado anteriormente. 

$\mathbf{\xi} = \begin{pmatrix}
1  \\
2 
\end{pmatrix}$.

$\mathbf{A \eta} = \mathbf{\xi} = \longrightarrow \big(\mathbf{A - I} \big)\mathbf{\eta} = \mathbf{\xi}$

$\big(\mathbf{A - I} \big)\mathbf{\eta} = \mathbf{\xi} \longrightarrow$
$\begin{pmatrix}
4 - 0& -2 \\
8 & -4 - 0
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
1 \\
2
\end{pmatrix} \longrightarrow$
$\begin{pmatrix}
4 & -2 \\
8 & -4
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
1 \\
2
\end{pmatrix}$

Resolvendo a equação, temos que $4 \eta_1 - 2 \eta_2 = 1 \longrightarrow \eta_2 = 2 \eta_1 - \frac{1}{2}$.

Se considerarmos $\eta_1$ como uma variável livre, com valor $k$, teremos:

$\mathbf{\eta} = \begin{pmatrix}
k \\
2k - \frac{1}{2}
\end{pmatrix} = \begin{pmatrix}
1 \\
2
\end{pmatrix}k + \begin{pmatrix}
0 \\
- \frac{1}{2}
\end{pmatrix}$.

Dos cálculos anteriores, temos:

$\mathbf{x}_2(t) = c_2 \Big( \begin{pmatrix}
1 \\
2
\end{pmatrix}t + \begin{pmatrix}
1 \\
2
\end{pmatrix} k + \begin{pmatrix}
0 \\
-\frac{1}{2}
\end{pmatrix} \Big)$

$\mathbf{x}(t) = \mathbf{x}_1(t) + \mathbf{x}_2(t) =  c_1 \begin{pmatrix}
1  \\
2 
\end{pmatrix} + c_2 \Big(  \begin{pmatrix}
1 \\
2
\end{pmatrix}t + \begin{pmatrix}
0 \\
-\frac{1}{2}
\end{pmatrix}  \Big) $

## Exercício 3 da página 436 do livro do Boyce, $10^{th}$ edition

### Descubra a solução geral para o seguinte sistema de equações diferenciais ordinárias de primeira ordem:

$\dot{\mathbf{x}} = \begin{pmatrix}
- \frac{3}{2} & 1 \\
 -\frac{1}{4} & -\frac{1}{2}
\end{pmatrix} \mathbf{x}$.

A solução geral é dada por:

$\mathbf{x}(t) = c_1 \mathbf{v}_1 e^{\lambda_1 t} + c_2 \mathbf{v}_2 e^{\lambda_2 t}$, onde $\lambda_1 e \lambda_2$ são os autovalores do sistema e $\mathbf{v}_1$ e $\mathbf{v}_2$ são os autovetores. Vamos provar que essa é uma solução do sistema.

Se $\mathbf{x} = \mathbf{v}e^{\lambda t}$ é uma solução de $\dot{\mathbf{x}} = \mathbf{Ax}$, 
$\dot{\mathbf{x}} = \lambda \mathbf{v}e^{\lambda t} =  \mathbf{Ax} = \mathbf{Av}e^{\lambda t} 
\longrightarrow \lambda \mathbf{v}e^{\lambda t} = \mathbf{Av}e^{\lambda t}$. Como $e^{\lambda t}$ jamais será $0.0$, podemos dividir ambos os lados da equação por $e^{\lambda t}$ e obteremos $\mathbf{Av} = \lambda \mathbf{v}$.

Calculando autovalores e autovetores:

$\mathbf{Ax} = \lambda\mathbf{x} \longrightarrow |(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = 0$

$|(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = \begin{vmatrix}
- \frac{3}{2} - \lambda & 1 \\
-\frac{1}{4} & -\frac{1}{2} - \lambda
\end{vmatrix}$

A equação característica é:

$\lambda^2 + 2 \lambda + 1 = 0 \longrightarrow \lambda = -1$. Só existe um autovalor.

Vamos calcular o autovetor associado a este autovalor.

$\begin{pmatrix}
- \frac{3}{2} & 1 \\
-\frac{1}{4} & -\frac{1}{2}
\end{pmatrix} 
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = (-1)
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} \longrightarrow \mathbf{v} =
\begin{pmatrix}
2 \\
1
\end{pmatrix}
$
A primeira parte da solução geral é:

$\mathbf{x}_1(t) = c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{-t}  $

Vamos calcular agora a segunda parte da equação geral, supondo que seja da seguinte forma:

$\mathbf{x}(t) = \mathbf{\xi} te^{\lambda t} + \mathbf{\eta} e^{\lambda t}$. Como $\lambda = -1$, temos
$\mathbf{x}(t) = \mathbf{\xi} te^{-t} + \mathbf{\eta}e^{-t}$.

$\dot{\mathbf{x}} = \mathbf{Ax}$

$\mathbf{\xi} e^{-t} - \mathbf{\xi} te^{-t} - \mathbf{\eta} e^{-t} = \mathbf{A \xi}t e^{-t} + \mathbf{A \eta}e^{-t}$

$ \mathbf{A \xi}t e^{-t} = -\mathbf{\xi} te^{-t} \longrightarrow \mathbf{A \xi} = -\mathbf{\xi}$. O vetor $\mathbf{\xi}$ é o autovetor de $\mathbf{A}$, já calculado anteriormente. 

$\mathbf{\xi} = \begin{pmatrix}
2  \\
1 
\end{pmatrix}$.

$\mathbf{\xi} e^{-t} - \mathbf{\eta} e^{-t} = \mathbf{A \eta}e^{-t} \longrightarrow \big(\mathbf{A + I} \big)\mathbf{\eta} = \mathbf{\xi}$

$\big(\mathbf{A + I} \big)\mathbf{\eta} = \mathbf{\xi} \longrightarrow$
$\begin{pmatrix}
-\frac{3}{2} + 1 & 1 \\
-\frac{1}{4} & -\frac{1}{2} + 1
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix} \longrightarrow$
$\begin{pmatrix}
-\frac{1}{2}& 1 \\
-\frac{1}{4} & \frac{1}{2}
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix}$

Resolvendo a equação, temos que $-\frac{1}{2} \eta_1 + \eta_2 = 2 \longrightarrow \eta_1 = 2 \eta_2 - 4$.

Se considerarmos $\eta_1$ como uma variável livre, com valor $k$, teremos:

$\mathbf{\eta} = \begin{pmatrix}
k \\
\frac{1}{2} + 2
\end{pmatrix} = \begin{pmatrix}
2 \\
1
\end{pmatrix}k + \begin{pmatrix}
0 \\
2
\end{pmatrix}$.

Dos cálculos anteriores, temos:

$\mathbf{x}_2(t) = c_2 \Big( \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{-t} + \begin{pmatrix}
2 \\
1
\end{pmatrix} k e^{-t} - \begin{pmatrix}
0 \\
2
\end{pmatrix}e^{-t} \Big)$

$\mathbf{x}(t) = \mathbf{x}_1(t) + \mathbf{x}_2(t) =  c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{-t} + c_2 \Big(  \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{-t} + \begin{pmatrix}
2 \\
1
\end{pmatrix} k e^{-t} - \begin{pmatrix}
0 \\
2
\end{pmatrix}e^{-t} \Big) = $

$\mathbf{x}(t) = c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{-t} + c_2 \Big(  \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{-t} +  \begin{pmatrix}
0 \\
2
\end{pmatrix}e^{-t} \Big) $

Vamos provar que o resultado obtido atende o sistema de equações dieferenciais.

$\dot{\mathbf{x}}(t) = -c_1 \begin{pmatrix}
2  \\
1 
\end{pmatrix}e^{-t} + c_2 \Big(  
\begin{pmatrix}
2 \\
1
\end{pmatrix}e^{-t} - \begin{pmatrix}
2 \\
1
\end{pmatrix}te^{-t}
+
\begin{pmatrix}
4 \\
0
\end{pmatrix}e^{-t} \Big) $

## Exercício 4 da página 436 do livro do Boyce, $10^{th}$ edition

### Descubra a solução geral para o seguinte sistema de equações diferenciais ordinárias de primeira ordem:

$\dot{\mathbf{x}} = \begin{pmatrix}
-3 & \frac{5}{2} \\
 -\frac{5}{2} & 2
\end{pmatrix} \mathbf{x}$.

A solução geral é dada por:

$\mathbf{x}(t) = c_1 \mathbf{v}_1 e^{\lambda_1 t} + c_2 \mathbf{v}_2 e^{\lambda_2 t}$, onde $\lambda_1 e \lambda_2$ são os autovalores do sistema e $\mathbf{v}_1$ e $\mathbf{v}_2$ são os autovetores. Vamos provar que essa é uma solução do sistema.

Se $\mathbf{x} = \mathbf{v}e^{\lambda t}$ é uma solução de $\dot{\mathbf{x}} = \mathbf{Ax}$, 
$\dot{\mathbf{x}} = \lambda \mathbf{v}e^{\lambda t} =  \mathbf{Ax} = \mathbf{Av}e^{\lambda t} 
\longrightarrow \lambda \mathbf{v}e^{\lambda t} = \mathbf{Av}e^{\lambda t}$. Como $e^{\lambda t}$ jamais será $0.0$, podemos dividir ambos os lados da equação por $e^{\lambda t}$ e obteremos $\mathbf{Av} = \lambda \mathbf{v}$.

Calculando autovalores e autovetores:

$\mathbf{Ax} = \lambda\mathbf{x} \longrightarrow |(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = 0$

$|(\mathbf{A} -\lambda\mathbf{I})\mathbf{x}| = \begin{vmatrix}
- 3 - \lambda & \frac{5}{2} \\
-\frac{5}{2} & 2 - \lambda
\end{vmatrix}$

A equação característica é:

$\lambda^2 + \lambda + \frac{1}{4} = 0 \longrightarrow \lambda = -\frac{1}{2}$. Só existe um autovalor.

Vamos calcular o autovetor associado a este autovalor.

$\begin{pmatrix}
-3 & \frac{5}{2} \\
-\frac{5}{2} & 2
\end{pmatrix} 
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} = (-\frac{1}{2})
\begin{pmatrix}
v_1 \\
v_2
\end{pmatrix} \longrightarrow \mathbf{v} =
\begin{pmatrix}

1
\end{pmatrix}
$
A primeira parte da solução geral é:

$\mathbf{x}_1(t) = c_1 \begin{pmatrix}
1  \\
1 
\end{pmatrix}e^{-\frac{1}{2}t}  $

Vamos calcular agora a segunda parte da equação geral, supondo que seja da seguinte forma:

$\mathbf{x}(t) = \mathbf{\xi} te^{\lambda t} + \mathbf{\eta} e^{\lambda t}$. Como $\lambda = -\frac{1}{2}$, temos

$\dot{\mathbf{x}} = \mathbf{Ax}$

$\mathbf{\xi} = \begin{pmatrix}
1  \\
1 
\end{pmatrix}$.

$\mathbf{\xi} e^{-\frac{1}{2}t} - \frac{1}{2}\mathbf{\eta} e^{-\frac{1}{2}t} = \mathbf{A \eta}e^{-\frac{1}{2}t} \longrightarrow \big(\mathbf{A + \frac{1}{2}I} \big)\mathbf{\eta} = \mathbf{\xi}$

$\big(\mathbf{A + \frac{1}{2}I} \big)\mathbf{\eta} = \mathbf{\xi} \longrightarrow$
$\begin{pmatrix}
-\frac{3}{2} + \frac{1}{2} & \frac{5}{2} \\
-\frac{5}{2} & 2 +\frac{1}{2}
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
1 \\
1
\end{pmatrix} \longrightarrow$
$\begin{pmatrix}
-\frac{5}{2}& \frac{5}{2} \\
-\frac{5}{2} & \frac{5}{2}
\end{pmatrix}
\begin{pmatrix}
\eta_1 \\
\eta_2
\end{pmatrix} = \begin{pmatrix}
1 \\
1
\end{pmatrix}$

Resolvendo a equação, temos que $-\frac{5}{2} \eta_1 + \frac{5}{2}\eta_2 = 1 $.

Se considerarmos $\eta_1$ como uma variável livre, com valor $k$, teremos:

\end{pmatrix} = \begin{pmatrix}

1
\end{pmatrix}k + \begin{pmatrix}
0 \\
\frac{2}{5}
\end{pmatrix}$.

\frac{2}{5}

 0 \\
